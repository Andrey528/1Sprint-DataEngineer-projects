Сгруппируйте вакансии по направлениями (DS, DE, Software Engenering, etc.)

```python
df.groupby('Title').count()
```

```python
df1 = df.loc[df['Salary'] != 'Не указана']

salary_dict = {}
index = df1.index
count = 0

for i in df1.Salary:
  salary_size = ''

  if 'до ' in i:
    i = i.split('до ')[1]

  if 'USD' in i:
    for j in i.split():
      if j.isdigit():
        salary_size += j
    i = int(salary_size) * 62.65

  else:
    for j in i.split():
      if j.isdigit():
        salary_size += j
    i = salary_size
  
  salary_dict[index[count]] = float(i)
  count += 1

df_salary = pd.DataFrame({'Salary':salary_dict})
```

Какая средняя и медианная зарплата по группам вакансий?

```python
df2 = pd.concat([df1['Title'], df_salary], axis=1)

df_out = df2.groupby('Title') ['Salary'].mean()

print(df_out)

df_out = df2.groupby('Title') ['Salary'].median()

print(df_out)
```

Какая средняя и медианная зарплата по каждому региону?

```python
df_region_salary = pd.concat([df1['Region'], df_salary], axis=1)

df_out = df_region_salary.groupby('Region') ['Salary'].mean()

print(df_out)

df_out = df_region_salary.groupby('Region') ['Salary'].median()

print(df_out)
```

Какая самая высокооплачиваемая из групп вакансий, исходя из их средних зарплат?

```python
df_out = df2.groupby('Title') ['Salary'].mean()
print(df_out[df_out == df_out.max()])
```

Какое процентное соотношение каждого региона по вакансиям от всех вакансий?

```python
x = df['Region'].value_counts(normalize=True) * 100
print(round(x, 2), '%', sep='')
```

```python
names = list(df[df['Sex'] == 'female'].Name)

for i in range(len(names)):
  if '(' in names[i]:
    names[i] = names[i].split('(')[1].replace(')', '')
  elif 'Miss.' in names[i]:
    names[i] = names[i].split('Miss. ')[1]
  elif 'Mrs.' in names[i]:
    names[i] = names[i].split('Mrs. ')[1]
  elif 'Ms.' in names[i]:
    names[i] = names[i].split('Ms. ')[1]
  elif 'Mlle.' in names[i]:
    names[i] = names[i].split('Mlle. ')[1]
  elif 'Mme.' in names[i]:
    names[i] = names[i].split('Mme. ')[1]

  if '"' in names[i]:
    names[i] = names[i].replace('"', '')
  print(names[i])
```

```python
names_dict = {}
for i in names:
  for j in i.split(' '):
    if j in names_dict:
      names_dict[j] += 1
    else:
      names_dict[j] = 1
```
